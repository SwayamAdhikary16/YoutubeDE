```python
import requests
import os
from dotenv import load_dotenv


# Load API Key from .env file
load_dotenv()
YOUTUBE_API_KEY = 'AIzaSyDJ-l2v6FPelZESqC5TUM_6ae5LCmkowtE'

# Your video ID
video_id = "sF5LYGgKbUA"

# YouTube API endpoint for video statistics
YOUTUBE_VIDEO_URL = "https://www.googleapis.com/youtube/v3/videos"

# API parameters
params = {
    "part": "statistics,snippet",
    "id": video_id,
    "key": YOUTUBE_API_KEY,
}

# Send GET request to fetch video stats
response = requests.get(YOUTUBE_VIDEO_URL, params=params)

# Check if the response is successful
if response.status_code == 200:
    data = response.json()
    
    if "items" in data and len(data["items"]) > 0:
        video_data = data["items"][0]
        stats = video_data["statistics"]
        
        print("Video Title:", video_data["snippet"]["title"])
        print("Channel:", video_data["snippet"]["channelTitle"])
        print("Views:", stats.get("viewCount", "N/A"))
        print("Likes:", stats.get("likeCount", "N/A"))
        print("Comments:", stats.get("commentCount", "N/A"))
    else:
        print("No video data available.")
else:
    print("Error:", response.status_code, response.text)

# YouTube API endpoint for comments
YOUTUBE_COMMENTS_URL = "https://www.googleapis.com/youtube/v3/commentThreads"

# API parameters for fetching comments
params_comments = {
    "part": "snippet",
    "videoId": video_id,
    "key": YOUTUBE_API_KEY,
    "maxResults": 5,
    "order": "time",
}

# Send GET request to fetch comments
response_comments = requests.get(YOUTUBE_COMMENTS_URL, params=params_comments)

# Check if the response for comments is successful
if response_comments.status_code == 200:
    data_comments = response_comments.json()
    
    if "items" in data_comments:
        print("\nRecent Comments:")
        for i, item in enumerate(data_comments["items"], start=1):
            comment = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
            print(f"{i}. {comment}")
    else:
        print("No comments found.")
else:
    print("Error:", response_comments.status_code, response_comments.text)
```
